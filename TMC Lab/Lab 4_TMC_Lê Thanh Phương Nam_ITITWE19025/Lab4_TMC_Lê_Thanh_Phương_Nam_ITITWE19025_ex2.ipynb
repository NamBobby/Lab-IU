```python
# Import Libs and time
import numpy as np
import matplotlib.pyplot as plt
import math as mt
from random import random
from random import seed
from mpl_toolkits.mplot3d import Axes3D
import timeit

#define the function
def f(x, y):
    return 3.5*x + 2*y + (x**2) - x**4 - 2*x*y - (y**2)

#create x, y array
x_axis = []
y_axis = []
x_gen_axis = []
y_gen_axis = []
```

```python
#define the random_search()
def random_search():
    # define initial x, y
    x=0.5
    y=0.5
    #cal f_initial
    f_old = f(x,y)
    eps_a = 1
    f_new = 1
    counter = 0
    while (abs(eps_a)>0.01):
        #range -2 to 2 because 0<random()<1
        x= (-2)+4*random()
        y= (-2)+4*random()
        f_new = f(x,y)
        x_gen_axis.append(x)
        y_gen_axis.append(y)
        #create counter
        counter += 1
        print("%s%20s%20s%20s%20s"%(counter, round(x,5), round(y,5), round(f_new,5),eps_a))
        if (f_new > f_old):
          eps_a = 1-f_new/f_old
          f_old = f_new
          x_axis.append(x)
          y_axis.append(y)
    return f_old

starttime = timeit.default_timer()
print(random_search())
print("The runtime is : ", timeit.default_timer() - starttime, 'second.')

z_axis = f(np.array(x_axis),np.array(y_axis))
fig = plt.figure()
fig.suptitle('All run points', fontsize = 20)
ax = plt.axes(projection = '3d')
ax.scatter3D(np.array(x_axis),np.array(y_axis),z_axis, c=z_axis,cmap='viridis',marker='o')
plt.show()

z_gen_axis = f(np.array(x_gen_axis),np.array(y_gen_axis))
fig = plt.figure()
fig.suptitle('All run points', fontsize = 20)
ax = plt.axes(projection = '3d')
ax.scatter3D(np.array(x_gen_axis),np.array(y_gen_axis),z_gen_axis, c=z_gen_axis,cmap='viridis',marker='o')
plt.show()
```
