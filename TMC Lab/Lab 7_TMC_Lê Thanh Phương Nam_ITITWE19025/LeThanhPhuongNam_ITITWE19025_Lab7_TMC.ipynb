```python
#Exercise 1:
import math # import lib
import timeit
from pylab import *
from scipy.special.orthogonal import p_roots
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
import scipy.integrate as integrate
from scipy.integrate import quad, romberg

%matplotlib inline

# define the function
def func1(x):
    return (x**(0.1))*(1.2 - x)*(1-np.exp(20*(x-1)))

X = np.arange(0,1,0.001)

starttime = timeit.default_timer()
#plot(X,f(X))
def F(x):
    res = np.zeros_like(x)
    for i,val in enumerate(x):
        # using gauss quadrature method
        y,err = integrate.quad(func1,0,val)
        res[i]=y
    return res
print ("Result: ", quad(func1,0,1))
plt.plot(X,F(X))
print("The runtime of exercise 1 is : ", round(timeit.default_timer() - starttime,4), 'second.')
print("Lê Thanh Phương Nam_ITITWE19025_Lab7TMC_ex1")
```
